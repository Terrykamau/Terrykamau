```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv("/content/survey lung cancer.csv")
data
```

```python
data.head()
```

```python
data['ANXIETY'].value_counts()
```

```python
label=LabelEncoder()
```

```python
labels=label.fit_transform(data['ANXIETY'])
```

```python
data['target']=labels
```
