```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,angles='xy',scale=1, scale_units='xy',color='g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
plt.quiver(0,0,4,5,angles='xy',scale=1, scale_units='xy',color='g')
plt.quiver(0,0,-3,-6,angles='xy',scale=1,scale_units='xy',color='b')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
vector1=np.asarray([0,0,2,3])
vector2=np.asarray([0,0,-1,2])
sum=vector1+vector2
plt.quiver(0,0,2,3,angles='xy',scale=1, scale_units='xy',color='g')
plt.quiver(0,0,-1,2,angles='xy',scale=1,scale_units='xy',color='b')
plt.quiver(0,0,1,5,angles='xy',scale=1, scale_units='xy',color='r')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
print(sum)
```

```python
vector1=np.asarray([0,0,4,3])
vector2=np.asarray([0,0,7,2])
subtraction=vector1-vector2
plt.quiver(0,0,4,3,angles='xy',scale=1, scale_units='xy',color='indigo')
plt.quiver(0,0,7,2,angles='xy',scale=1,scale_units='xy',color='orange')
plt.quiver(0,0,-3,1,angles='xy',scale=1, scale_units='xy',color='black')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
print(subtraction)
```

```python
vector1=np.asarray([0,0,5,9])
vector2=np.asarray([0,0,7,-4])
multiplication=vector1*vector2
plt.quiver(0,0,5,9,angles='xy',scale=1,scale_units='xy',color='purple')
plt.quiver(0,0,7,-4,angles='xy',scale=1,scale_units='xy',color='yellow')
plt.quiver(0,0,35,-36,angles='xy',scale=4,scale_units='xy',color='red')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

DOT PRODUCT AND CROSS PRODUCT

```python
a= np.array([6,-8])
b= np.array([3,4])
dot=np.dot(a,b)
print(dot)
```

```python
a= np.array([6,-8])
b= np.array([3,4])
cross=np.cross(a,b)
print(cross)
```

```python
a= np.array([6,-8,9])
b= np.array([3,4,5])
cross=np.cross(a,b)
print(cross)
```

```python
a=np.array([6,9])
b=np.array([7,6])
magnitude_b=np.sqrt(sum(b**2))
projection=(np.dot(a,b)/magnitude_b)*b
print(magnitude_b)
print(projection)
```

```python
a=np.array([6,9,7])
b=np.array([7,6,6])
magnitude_b=np.sqrt(sum(b**2))
projection=(np.dot(a,b)/magnitude_b)*b
print(magnitude_b)
print(projection)
```
